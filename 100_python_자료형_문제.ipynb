문제3)   
'spam ham'을 'ham spam'으로 치환하세요.   

※문자열의 split( )과 join( )메서드 그리고 리스트의 reverse( )메서드를 사용

```python
s = 'spam ham'
a = s.split(' ')
b = a[-1::-1]   # a.reverse() 이거 쓰면 한번에 된다
c = ' '.join(b)
print(c)
```

문제6)  
100  이하의 자연수 중에서 2의 배수이거나 3의 배수인 수의 개수를 구하세요.

[실행결과]  
67

```python
s = []

for i in range(1,101):
    if i%2 ==0 and i%3 ==0:
        s.append(i)
print(len(s))
```
