## 예외(exception)
- 코드를 실행하는 중에 발생한 오류이다.
- 오류가 발생하는 이유는 프로그램이 잘못 동작하는 것을 막기 위함이다.


## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy


````python
a = 10 / 0
````

````python
# try 블록이 정상적으로 실행되지 않으면 else블록은 실행되지 않는다.
# try영역에서 예외가 발생했을 때: try 실행 -> except 실행 -> finally 실행
# try영역이 정상적으로 수행했을 때: try실행 -> else 실행 -> finally 실행

try:
    a = int(3)
    b = int('a')  # 여기가 문제이다. int값을 못받는다.
    c = int(4)

except:                 # 예외가 발생한다.              # 밑에 3개중에 한개를 무조건 써야 한다. try만 쓰면 안된다
    print('숫자가 아닌 문자가 입력되었습니다')
else:
    print(f'{a}와{b}와{c}의 합:{a+b+c}')
finally:               # 무조건 수행하는 코드
    print('program end')
````

## 다중 except
- 다중 except 선언이 가능하다.
- except뒤에 오는 as는 try 블록에서 발생한 이름을 받아온다.
```
  try:
     실행할 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드  
  except:
     예외가 발생했을 때 처리하는 코드 


````python
y = [10, 20, 30]

try:      # 1 4
    index , x = map(int, input('인덱스와 나눌 숫자를 입력하세요.').split())
    print(y[index] // x)
except ZeroDivisionError as e: 
    print('숫자를 0으로 나눌 수 없습니다', e)
except IndexError as e:
    print('잘못된 인덱스입니다.', e)
except Exception as e:    # 어떤 에러가 발생할지 예측할 수 없을 때
    print(e)
else:
    print('정상처리 되었습니다')
finally:
    print('program end')
````

````python
# 이건 뭔 내용인지 이해가 안되네.....

class Bird:
    def fly(self):
        raise NotImplementedError # 상속받은 자식클래스에서 fly()을 오버라이딩 해줘야 한다.
    
class Eagle(Bird):

    def display(self):
        print('display')

    def fly(self):        # fly() 오버라이딩을 안하면 오류 발생   # 일단 오버라이딩 이라는 개념을 좀 알아놔야겠다..
        print('very fast')

eagle = Eagle() 
eagle.fly()   
````

````python
from abc import abstractmethod


class Bird:
  @abstractmethod   # -> 이게 뭐냐..... 그냥 모르겠다..
  def fly(self):
    pass


class Eagle(Bird):
  def display(self):
    print('display')


#   def fly(self):
#   print('very fast')


eagle = Eagle()
eagle.fly()
````

````python
from abc import ABC, abstractmethod


class Bird(ABC):
  @abstractmethod   
  def fly(self):
    pass


class Eagle(Bird):
  def display(self):
    print('display')


#   def fly(self):
#   print('very fast')

# 오버라이딩을 안하면 생성단계에서 오류가 생긴다. / 추상 클래스는 객체 생성을 할 수 없다...
eagle = Eagle()   # ABC을 상속받은 Bird추상클래스이다. Bird을 상속받은 Eagle도 추상클래스이다.
eagle.fly()
````

## 사용자예외클래스 만들기

````python
class UserClass(Exception):
    def __str__(self):
        return '허용하지 않는 데이터입니다'
````

````python
try:
    x = int(input('숫자입력:'))   # 숫자가 아니라 문자를 입력하니 완전히 오류가 떴다.
    if x <= 0:
        raise UserClass
    print(x)

# 위로 갈 수록 구체적인 예외 종류들이 와야하고, 아래로 갈수록 추상적인 예외 종류 들이 와야한다.
except UserClass as e:
    print(e)
except ValueError as e:    # 문자로 입력하면 Value error이기 때문에 이것만 추가하면 된다.
    print(e)
except Exception as e:  # 여기서 다 걸러진다. 그렇게 되면 밑의 친구들이 의미가 없어진다. 그래서 아래로 빼주는 거다
    print(e)            # 다중 except을 사용할 때 부모클래스는 아래에 위치한다.  
except BaseException as e:
    print(e)
````
